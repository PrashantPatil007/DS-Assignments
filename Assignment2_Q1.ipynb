```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("D:\DS Assignments\Assignments\Assignment 2\Assignment2_Q1.csv")
```

```python
data.head()
```

```python
column1 = data['Name of company']
column2 = data['Measure X'].str.rstrip('%').astype('float') / 100.0

# Create a bar plot
plt.bar(column1, column2)

# Set the title and labels for x and y axes
plt.title('Bar Chart of Two Columns')
plt.xlabel('Column 1')
plt.ylabel('Column 2')
plt.yticks([i/10.0 for i in range(11)], ['{}%'.format(i*10) for i in range(11)])

```

```python
plt.hist(column2, bins=10, edgecolor='black')

# Set the title and labels for x and y axes
plt.title('Histogram of Column 2')
plt.xlabel('Column 2 (Percentage)')
plt.ylabel('Frequency')

# Set the x-axis ticks to show percentages
plt.xticks([i/10.0 for i in range(11)], ['{}%'.format(i*10) for i in range(11)])
```
